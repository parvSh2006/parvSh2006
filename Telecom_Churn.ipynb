## Calculate Correlation Matrix

### Subtask:
Compute the pairwise correlation between all numerical columns in the DataFrame `df`.


**Reasoning**:
To compute the pairwise correlation between all numerical columns, I will use the `.corr()` method on the DataFrame `df` and store the result in a variable named `correlation_matrix`.

**Reasoning**:
To visualize the correlation matrix, I will use `seaborn.heatmap` with annotations enabled. This will provide a clear graphical representation of the pairwise correlations between all numerical columns.



```python
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
numerical_cols = df.select_dtypes(include=np.number).columns

outliers_data = {}

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    if not outliers.empty:
        outliers_data[col] = {
            'count': len(outliers),
            'percentage': (len(outliers) / len(df)) * 100
        }

print("Outlier Detection Results (IQR Method):")
for col, data in outliers_data.items():
    print(f"Column '{col}': {data['count']} outliers found ({data['percentage']:.2f}% of data)")
```

## Summarize Key Churn Drivers and Predictors

### Subtask:
Based on the analyses of tenure, price, and services, synthesize the findings to determine which type of customer churns the most, what drives churn (price, tenure, services), and identify the most likely strong predictor features for churn.

#### Summary of Churn Drivers and Predictors:

**1. Overall Churn Rate:**
*   Approximately 14.5% of customers churn, indicating that churn is a significant issue, though it represents a minority of the customer base.

**2. Contract Renewal (`ContractRenewal`):**
*   **Key Finding:** Customers who do *not* renew their contract (ContractRenewal=0) have a significantly higher churn rate (42.4%) compared to those who renew (11.5%).
*   **Implication:** Lack of contract renewal is a very strong predictor of churn. Customers on month-to-month plans or nearing the end of their contract are highly prone to churning.

**3. Customer Service Calls (`CustServCalls`):**
*   **Key Finding:** Customers who churn make a higher average number of customer service calls (mean ~2.23) compared to non-churning customers (mean ~1.45). The count plot clearly shows an increase in churn percentage with an increasing number of customer service calls.
*   **Implication:** High customer service call volume is a strong indicator of dissatisfaction and a driver of churn.

**4. Data Plan (`DataPlan`) and Data Usage (`DataUsage`):**
*   **Key Finding (DataPlan):** Customers *without* a data plan (DataPlan=0) have a higher churn rate (16.7%) than those *with* a data plan (8.7%).
*   **Key Finding (DataUsage):** Counter-intuitively, non-churning customers have a slightly higher average data usage (mean ~0.86) than churning customers (mean ~0.55). This suggests that customers who actively use data services (and thus likely have a data plan) are more engaged and less likely to churn.
*   **Implication:** Having a data plan, and likely using it, contributes to customer retention. The absence of a data plan makes customers more susceptible to churn.

**5. Daily Minutes Used (`DayMins`) and Daily Calls Made (`DayCalls`):**
*   **Key Finding (DayMins):** Churning customers have a noticeably higher average daily minute usage (mean ~206.9 minutes) compared to non-churning customers (mean ~175.2 minutes). The box plot shows a higher median and wider spread for churned customers.
*   **Implication:** Customers with very high daily minute consumption might be more prone to churn, potentially due to seeking better rates elsewhere for heavy usage.
*   **Key Finding (DayCalls):** There is no significant difference in the number of daily calls made between churned and non-churned customers.
*   **Implication:** `DayCalls` is not a strong predictor of churn.

**6. Monthly Charge (`MonthlyCharge`):**
*   **Key Finding:** Churning customers have a slightly higher average monthly charge (mean ~59.19) than non-churning customers (mean ~55.82). The box plot shows that churned customers tend to be in the upper quartile of monthly charges.
*   **Implication:** While not as strong as contract renewal or customer service calls, higher monthly charges can contribute to dissatisfaction and drive churn.

**7. Account Weeks (`AccountWeeks` - Tenure):**
*   **Key Finding:** The descriptive statistics and box plot for `AccountWeeks` show very similar distributions for both churned and non-churned customers (mean of ~100.8 for non-churned vs. ~102.7 for churned). There is no substantial difference in tenure between the two groups.
*   **Implication:** Customer tenure alone does not appear to be a strong direct predictor of churn in this dataset. Both new and long-term customers churn at similar rates.

**8. Overage Fee (`OverageFee`) and Roaming Minutes (`RoamMins`):**
*   **Key Finding (OverageFee):** Churning customers have a slightly higher average overage fee (mean ~10.62) compared to non-churning customers (mean ~9.95).
*   **Key Finding (RoamMins):** Churning customers have a slightly higher average roaming minutes (mean ~10.70) compared to non-churning customers (mean ~10.16).
*   **Implication:** While the differences are subtle, higher overage fees and roaming usage might contribute to a perception of higher costs and thus to churn, but they are not as impactful as other factors.

### Conclusion:

**Type of customer most prone to churn:**
Customers most prone to churn are those who have **not renewed their contract**, make an **above-average number of customer service calls**, do **not have a data plan**, and exhibit **high daily minute usage**. They also tend to face slightly higher monthly charges and potentially incur more overage fees or roaming charges.

**Main drivers of churn:**
1.  **Lack of Contract Renewal:** This is the most significant driver, indicating customers without a contractual commitment are highly volatile.
2.  **Poor Customer Experience/Service Issues:** Evidenced by a higher number of customer service calls.
3.  **High Usage Costs:** High `DayMins` (leading to potential high bills) and slightly elevated `MonthlyCharge` and `OverageFee` are contributing factors.
4.  **Lack of Service Engagement/Value:** Customers without a data plan seem less engaged and more likely to churn.

**Most likely strong predictor features for churn:**
1.  **`ContractRenewal`** (inverse relationship, i.e., no renewal leads to high churn)
2.  **`CustServCalls`** (positive relationship, more calls lead to high churn)
3.  **`DayMins`** (positive relationship, higher usage leads to high churn)
4.  **`DataPlan`** (inverse relationship, i.e., no data plan leads to high churn)
